```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import matplotlib.cbook as cbook


with cbook.get_sample_data('C:\\Users\\HP\\Desktop\\ninhbinh.jpg') as image_file:
    image = plt.imread(image_file)

fig, ax = plt.subplots()
im = ax.imshow(image)
patch = patches.Circle((500, 500), radius=300, transform=ax.transData)
im.set_clip_path(patch)

ax.axis('off')
plt.show()
```
